- Object detection is the task of finding the different objects in an image and classifying them.

### Understanding R-CNN

- The goal of R-CNN is to take in an image, and correctly identify where the main objects (via a bounding box) in the image.

    - **Inputs: Image**
    - **Outputs: Bounding boxes + labels for each object in the image.**

- But how do we find out where these bounding boxes are?

 - **Propose a bunch of boxes in the image and see if any of them actually correspond to an object.**

![](assets/understanding.png)

- R-CNN creates these bounding boxes, or region proposals, using a process called **Selective Search** (http://www.cs.cornell.edu/courses/cs7670/2014sp/slides/VisionSeminar14.pdf)

- At a high level, Selective Search looks at the image through windows of different sizes, and for each size tries to group together adjacent pixels by:
    - texture
    - color
    - or intensity

![](assets/rcnn-2.png)

- In step 3, R-CNN passes the warped region to a modified version of **AlexNet**
- In step 4, on the final layer fo the CNN, R-CNN adds a **Support Vector Machine(SVM)** to classify whether this is an object, and if so what object.



### Improving the Bounding Boxes

- To tighten the bounding box to fit the true dimensions fo the object, we run a **simple linear regressionn** on the region proposal. 

- Here are the inputs and outputs of this regression model:

    - **Inputs:** sub-regions of the image corresponding to objects.
    - **Outputs:** New bounding box coordinates for the object in the sub-region.


**So, to summarize, R-CNN is just the following steps:**

1. Generate a set of proposals for bounding boxes.

2. Run the images in the bounding boxes through a pre-trained AlexNet and finally an SVM to see what object the image in the box is.

3. Run the box through a linear regression model to output tighter coordinates for the box once the object has been classified.

![](assets/rcnn-conv.png)
Image source: http://cs231n.stanford.edu/slides/2017/cs231n_2017_lecture11.pdf

### Fast R-CNN Insight 1: RoI(Region of Interest) Pooling

- Just run the CNN just once per image and then find a way to share that computation across the ~ 2000 proposals

![](assets/roi.png)

- So, all it takes us is one pass fo the original image to generate region proposals as opposed to ~2000.


### Fast R-CNN Insight 2: Combine All models into One Network

![](assets/fast-rcnn-1.png)

- Jointly train the CNN, classifier, and bounding box regressor in a single model. R-CNN uses different models to extract iamge features (CNN), classify (SVM), and tighten bouding boxes (regressor), **Fast R-CNN instead uses a single network to compute all three.**

![](assets/fast-rcnn.png)
Image source: http://cs231n.stanford.edu/slides/2017/cs231n_2017_lecture11.pdf

### Class Names

The model classifies objects and returns class IDs, which are integer value that identify each class. Some datasets assign integer values to their classes and some don't. For example, in the MS-COCO dataset, the 'person' class is 1 and 'teddy bear' is 88. The IDs are often sequential, but not always. The COCO dataset, for example, has classes associated with class IDs 70 and 72, but not 71.

To improve consistency, and to support training on data from multiple sources at the same time, our Dataset class assigns it's own sequential integer IDs to each class. For example, if you load the COCO dataset using our Dataset class, the 'person' class would get class ID = 1 (just like COCO) and the 'teddy bear' class is 78 (different from COCO). Keep that in mind when mapping class IDs to class names.

To get the list of class names, you'd load the dataset and then use the class_names property like this.

```python
# Load COCO dataset
dataset = coco.CocoDataset()
dataset.load_coco(COCO_DIR, "train")
dataset.prepare()

# Print class names
print(dataset.class_names)
```
